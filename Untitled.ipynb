```python
from matplotlib import pyplot as plt
```

```python
import pickle
```

```python
with open('result/train_loss.pickle','rb') as f:
    train_loss = pickle.load(f)
with open('result/valid_loss.pickle','rb') as f:
    valid_loss = pickle.load(f)
train_loss = [i/32963*128 for i in train_loss]
```

```python
fig, ax = plt.subplots()
plt.plot(range(len(train_loss)),train_loss)
plt.plot(range(len(valid_loss)),valid_loss)
ax.legend(['train_loss', 'valid_loss'])
plt.savefig('result/loss_graph.png')
```
